# EmbolcallRNAseq Tutorial


We will demonstrate the EmbolcallRNAseq pipeline by analyzing a bulk RNA seq time series data derived from mouse tumors. The experiment was published in and involves the collection of microtumors from the STAMP tumor model as well as healing pores. A healthy skin sample was added as a control for the unpeturbed tissue. The goal of the experiment was to study which are the factors that are different in the pore healing process with the presence or absence of tumor cells. The pipeline put forward in this paper, will provide the tools to interpret the biology of the signals present in this transcriptomic dataset.

## Pre-Processing

```r
########################## 1. Set-up
# Libraries
library(EmbolcallRNAseq)
library(edgeR)
library(PCAtools)
library(Biobase)
library(tidyverse)
library(ggpubr)
# Working Directory
setwd("/Users/pechuanj/Documents/Workstation/stamp/NGS3947_WoundHealing/")
# Gene Sets
species = "Mus musculus"
ListSupport(c(wp2gene,m_h,m_kegg,m_c2,m_c5,m_c7,dbGSEA,cellMarkers,keggdb),LoadGSEADB(species))
```

The data is loaded as an DGEList object which allows us to interface with edgeR-limma pipelines to posteriorly analyze the differentially expressed genes. The DGEList contains the count matrix and the sample annotations.

```r
################### 2. Loading and prepare data
# Load ESET 
es.all = readRDS("./data/NGS3947_ESET.Rdata")
# Prepare DGEList 
y.all = DGEList(
  exprs(es.all),genes=fData(es.all))
y.all$samples = cbind(y.all$samples, pData(es.all))
y.all = calcNormFactors(y.all)
pheno = yf$samples
y.all %>% head()
```

As a first step, we filter lowly expressed genes which tend to obscure the signal of the data an impact our type I error correction process by adding variables that are not interesting. These also interfere with the voom-limma pipeline and thus is an important pre-processing step. We will filter the genes that have a minimum logcpm in the largest sample size coming from the experimental group that we are analysing, in this case "Day". We then calculate normalization factor with edgeR using the TMM method and the resulting log transformed matrix of normalized expression will be the input of our exploratory analyses.

```r
############################ 3. Filter lowly expressed genes
keep.exprs = rowSums(edgeR::cpm(es.all) >= 0.5) > max(table(pheno$Day))
yf =  y.all[keep.exprs,, keep.lib.sizes=FALSE]
yf = calcNormFactors(yf,method="TMM")
# Expression matrices
ExprMat = cpm(yf,log=T) # After calling edger::calcNormFactors() is TMM normal
head(ExprMat)
```

We define a color schema for the sample annotations we want to consider as co-variates. In this case, we want to track tissue type and the day it was sampled and a combined annotation.

```r
############################ 4. Obtain Annotations
#  Get the covariates to analyze
annotations = pheno[,c("Complete","Tissue","Day")]
# Aesthetics
CompletePalette = c("firebrick1","firebrick3","firebrick4","chocolate","green1","green3","green4")
TissuePalette = c("firebrick4","chocolate","green4")
DayPalette = c("dodgerblue1","dodgerblue3","dodgerblue4")

list_palettes = list(CompletePalette,TissuePalette,DayPalette)
top_bar=PrepareTopBar(annotations=annotations,list_palettes=list_palettes)
annotations %>% head()
```

## Exploratory Analyses


Before digging into the differential expression analyses and pathway interpretation, we first look in an unsupervised manner at the data. To do this, we first select the most variable genes from expression matrix and perform a hierachical clustering on the resulting matrix. The number of variable genes to consider can be determined using projection score.  We annotate the resulting heatmap by looking at the gene clusters and their annotations based on GO terms or other ontology databases. 

```r
########################### 5. Unsupervised Hierachical Clustering on the Most Variable Genes
# Subset the expression matrix to most variable genes
nvar = 200 # Via Projection score
topdat = NthMostVariableFeatures(ExprMat,nvar) 

# Row Scale the matrix
scaled_mat = t(scale(t(topdat)))

# Annotate by Go Terms Go
GOHeatMap(scaled_mat,Gomet = "GO",nk_Go = 10,organismRef = "Mus musculus",nHL = 5,Csplit=6,
              title = paste0("Top ",nvar," Most Variable Genes"),top_annotation = top_bar)
```

We can see that annotating the heatmap by Wiki pathways reveals that the wound healing pathway genes go down in the healthy skin and the last timepoints of the pore samples. The tumor samples, on the contrary, develop an interferon gamma signature consistent with lymphocyte recruitment.

Another avenue for exploratory analysis is the use of dimensionality reductions methods. We refer to the package PCAtools as a great set of functions to perform these analyses. 

```r
################################## 6. PCA on the same set
# Quick exploration
p =  PCAtools::pca(scaled_mat, metadata = pheno, removeVar = 0)
# BiPlot to explore the loadings
biplot(p, showLoadings = T,
       labSize = 3, pointSize = 2, 
       sizeLoadingsNames = 3,
       colLoadingsNames = 'red4',
       fillBoxedLoadings = alpha("white", 2/4),
       colby="Complete",
       colkey =  CompletePalette,
       ntopLoadings=5,
       lab = NULL,
       alphaLoadingsArrow=0.8,
       colLegendTitle = 'Sample:',
       # Ellipse
       ylim = c(-20,20),
       xlim = c(-25,25),
       ellipse =  T,
       hline = 0, vline = c(0),
       title = paste0("PCA of Top ",nvar," Variable Genes"),
       legendPosition = 'top', 
       legendLabSize = 10, 
       legendIconSize = 8.0)
```

## Cell type and pathway scoring

Before conducting the differential expression analysis and its posterior interpretation, it would be useful for the interpretation of the outcome to score our data with a few signatures of interests. In this dataset, we set ourselves to explore some inflammatory signatures like interferon related genes. We can do this using a score computed by taking the first principal compotnents of the data matrix restricted to the genes that make-up the signature.

```r
################################## 7. Signatures of Interests
# Interferon
inf = c("Rsad2","Mx1","Mx2","Irf7","Irf9","Cxcl9","Cxcl10","Mb21d1","Oasl2","Oas1b","Ifit1","Ifit2","Ifit3","Ifitm1",
        "Ifitm2")
# Restrict the expression matrix
ExprMatIfn = ExprMat[(rownames(ExprMat) %in% inf),]

# G-Score
pcSig = gsScore(ExprMatIfn)
yf$samples$pcSigJeremie = pcSig
scores = yf$samples
scores$Complete = scores$Complete  %>% factor() %>% forcats::fct_recode("Skin"="SkinDay8")
scores$Complete = factor(scores$Complete,levels = c("Skin",
                                                    "TumorDay1",
                                                    "PoreDay1",
                                                    "TumorDay3", 
                                                    "PoreDay3",
                                                    "TumorDay8",
                                                    "PoreDay8"))

# Plot
my_comparisons = list(c("TumorDay1","Skin") ,
                      c("TumorDay3","Skin"),
                      c("TumorDay8","Skin"),
                      c("PoreDay1","Skin"),
                      c("PoreDay3","Skin"),
                      c("PoreDay8","Skin"))


ggboxplot(scores, x = "Complete", y = "pcSigJeremie",
          color = "Complete", palette = CompletePalette ,add="jitter")+
 stat_compare_means(comparisons = my_comparisons,method ="t.test",
                     method.args = list(alternative = "two.sided"),
                    p.adjust.method = "fdr",var.equal=F,
                   label="p.signif")+
  ylab("Interferon Response Gene Score")+
  xlab("")+
  theme(axis.text.x = element_text(angle = 90,hjust = 1),legend.position = "none")+
  font("x.text", size = 11)+font("y.text", size = 11)
```

For this experiment, we also had a companion single cell RNA sequencing experiment, we collected a set of specific gene sigtaures that allow us to identify the different cell populations comprising our tumors. We can perform a deconvolution using our principal component scores based on this gene list. This can be potentially used to explore other datasets. Our function, however, accomodates for a user defined list.

```r
################################# 8. ATLAS interpretation
pure_markers=read.csv("~/Downloads/sig_purify_genes.csv",header = TRUE)

AtlasScoreHeatmap(ExprMat,pure_markers,Rsplit = 4,Csplit = 4) 


```

The deconvolution consitently identifies the lymphoid cells and increased vasculature characteristic of tumors in late timepoints. 

## Differential expression

```r
Treatment = factor(pheno$Complete)
design = model.matrix(~0+Treatment)
colnames(design) = levels(Treatment)
# Run voom 
vmf=voomWithQualityWeights(yf, design, plot=T)
```

```r
# Run Limma
fit=lmFit(vmf, design)
# Find differentially expressed genes
contrast.matrix = makeContrasts("Day1" = TumorDay1 - PoreDay1,
                                "Day3" = TumorDay3 - PoreDay3,
                                "Day8" = TumorDay8 - PoreDay8,
                                 levels=design)

fit2 = contrasts.fit(fit,contrast.matrix)
fit2 = eBayes(fit2)
dt = decideTests(fit2,lfc = 0,p.value = 0.05)
summary(dt)
vennDiagram(dt)
```
